Define Y (target variable) and X (Features)

```python
y = wine["quality"]
x= wine[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

```python
y.shape
```

```python
x.shape
```

```python
y
```

```python
x
```

Standardizing X variable

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

Getting Traing-Test-Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_Train, x_Test, y_Train, y_Test = train_test_split(x,y, random_state=2529)
```

```python
x_Train.shape, x_Test.shape , y_Train.shape, y_Test.shape
```

Get Model Train

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit( x_Train , y_Train)
```

Getting Model Prediction

```python
y_pred = model.predict(x_Test)
```

```python
y_pred.shape
```

```python
y_pred
```

Getting Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_Test, y_pred))
```

```python
print(classification_report(y_Test, y_pred))
```

Rerunning model with two class created for wine quality

```python
y = wine["quality"].apply(lambda y_value: 1 if y_value >=6 else 0)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_Train, x_Test, y_Train, y_Test = train_test_split(x,y, random_state=2529)
```

```python
x_Train.shape, x_Test.shape, y_Train.shape, y_Test.shape
```

```python
model.fit(x_Train, y_Train)
```

```python
y_pred = model.predict(x_Test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
print(confusion_matrix(y_Test, y_pred))
```

```python
print(classification_report(y_Test, y_pred))
```
